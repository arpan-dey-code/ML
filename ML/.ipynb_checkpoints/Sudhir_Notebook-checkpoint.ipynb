# ML06  :  Python Program to Compute all the Permutation of the String(Using Recursion)

```python
def get_permutations(string, i=0):
    if i==len(string):
        print(" ".join(string))
    for j in range(i,len(string)):
        words= [c for c in string]
        words[i],words[j]= words[j], words[i]     
        get_permutations(words,i+1)
get_permutations("JAY")   
```

# ML04: Python Program to print Fibonacci Series:(using for loop):-

```python
def fib(n):
    a=0
    b=1
    print(a)
    print(b)
    
    for i in range(2,n):
        c=a+b
        a=b
        b=c
        print(c)
fib(5)        
    
```

# ML05: Python Program to print Fibonacci Series:(using WHILE loop):-

# MLO4: Python program to determine whether a given year is a leap year or Not:-

```python
year=int(input("Enter a year:-"))
if(year%4==0  or year%400==0 and year%100!=0 ):
        print("The year is a leap year:-")
else:    
        print("The year is not a leap year")
```

```python
def print_leap_years(lower, upper):
    leap_years = [year for year in range(lower, upper + 1) if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)]
    
    if leap_years:
        print("Leap years between {} and {}: {}".format(lower, upper, leap_years))
    else:
        print("No leap years between {} and {}.".format(lower, upper))


    
```

# ML02: Python program to print sum of numbers using for loop

```python
sum=0

print("The first 10 natural numbers are:-")
for i in range(1,11):
    print(i,end=" ")
    sum=sum+i
print()   
print("The sum of numbers is:-",sum)
```

## ML 01: Python program to input number and sum of its digits by using while loop:

```python
n= int(input("Enter Number:"))
sum=0
num=n
while (n>0):
    r=n%10
    sum= sum+r
    n=n//10
    print("The Sum of digits of the number ",num," is :",sum)
```
