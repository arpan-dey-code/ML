```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Number of data points
num_data_points = 1000

# Generate real-world data for features
advertising_spend = np.random.uniform(1000, 10000, size=num_data_points)
store_visitors = np.random.randint(100, 1000, size=num_data_points)
average_temperature = np.random.uniform(50, 100, size=num_data_points)
promotion_indicator = np.random.choice([0, 1], size=num_data_points)
competition_distance = np.random.uniform(1, 10, size=num_data_points)
customer_satisfaction_rating = np.random.randint(1, 11, size=num_data_points)
inventory_level = np.random.randint(100, 1000, size=num_data_points)

# Generate sales data based on features with adjusted coefficients and increased noise
sales = 30 * advertising_spend + 10 * store_visitors + 5 * average_temperature + 20 * promotion_indicator + 8 * competition_distance + 6 * customer_satisfaction_rating + 0.8 * inventory_level + np.random.randn(num_data_points) * 10000  # Increased noise

# Create a DataFrame to store the data
data = pd.DataFrame({
    'Advertising Spend': advertising_spend,
    'Store Visitors': store_visitors,
    'Average Temperature': average_temperature,
    'Promotion Indicator': promotion_indicator,
    'Competition Distance': competition_distance,
    'Customer Satisfaction Rating': customer_satisfaction_rating,
    'Inventory Level': inventory_level,
    'Sales': sales
})

# Save the data to a CSV file
data.to_csv('Store Sales.csv', index=False)

# Scatter plot with adjusted dot size
plt.figure(figsize=(10, 6))
plt.scatter(data['Sales'], data['Advertising Spend'], color='blue', s=1, label='Advertising Spend')
plt.scatter(data['Sales'], data['Store Visitors'], color='green', s=1, label='Store Visitors')
plt.scatter(data['Sales'], data['Average Temperature'], color='orange', s=1, label='Average Temperature')
plt.scatter(data['Sales'], data['Promotion Indicator'], color='red', s=1, label='Promotion Indicator')
plt.scatter(data['Sales'], data['Competition Distance'], color='purple', s=1, label='Competition Distance')
plt.scatter(data['Sales'], data['Customer Satisfaction Rating'], color='brown', s=1, label='Customer Satisfaction Rating')
plt.scatter(data['Sales'], data['Inventory Level'], color='pink', s=1, label='Inventory Level')
plt.title('Scatter Plot of Sales vs. Features')
plt.xlabel('Sales')
plt.ylabel('Features')
plt.legend()
plt.grid(True)
plt.show()
```
